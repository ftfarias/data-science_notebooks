```python
def func(x, a, b, c):
    return a*np.exp(-b*x) + c

x = np.linspace(0,4,50)
y = func(x, 2.5, 1.3, 0.5)
yn = y + 0.2*np.random.normal(size=len(x))

plt.figure()
plt.plot(x,y,"r-")
plt.plot(x,yn,"b-")
plt.title(u'Gráfico Maravilhoso !')
plt.show()
```

```python
def func(x, a, b, c):
    return a*np.exp(-b*x) + c

x = np.linspace(0,4,50)
y = func(x, 2.5, 1.3, 0.5)
yn = y + 0.2*np.random.normal(size=len(x))

with plt.xkcd():
    plt.figure()
    plt.plot(x,y,"r-")
    plt.plot(x,yn,"b-")
    plt.title(u'Gráfico Maravilhoso !')
    plt.show()
```
