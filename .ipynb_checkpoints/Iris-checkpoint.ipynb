```python
iris = datasets.load_iris()
```

```python
d = pd.concat([pd.DataFrame(iris.data),pd.DataFrame(iris.target)],axis=1)
d.columns = iris.feature_names + ['classe']
```

```python
d.head()
```

```python
d.describe()
```

```python
d.info()
```

```python
# é necessário converter classe para object para the o ggplot considere como um fator, e nao uma escala, com cores erradas
d.classe = d.classe.astype(np.object)
d.info()
```

```python
d.to_csv("iris.csv")
```

```python
d.classe.unique()
```

```python
ggplot(aes(x=d.columns[0], y=d.columns[1], color='classe', shape='classe'), data=d ) +\
    geom_point() +\
    scale_color_brewer(type='qual') +\
    xlab(d.columns[0]) + ylab(d.columns[1]) + ggtitle("Iris")
```

```python
#f = plt.figure(figsize=(15, 6))
def getByClass(classe,idx1,idx2):
    x = d[d['classe'] == classe][d.columns[idx1]].values
    y = d[d['classe'] == classe][d.columns[idx2]].values
    return (x,y)
x,y = getByClass(0,0,1)
pl.plot(x,y,'o',color="r")
x,y = getByClass(1,0,1)
pl.plot(x,y,'o',color="b")
x,y = getByClass(2,0,1)
pl.plot(x,y,'o',color="g")
pl.show()
```

```python
x,y = getByClass(0,1,2)
pl.plot(x,y,'o',color="r")
x,y = getByClass(1,1,2)
pl.plot(x,y,'o',color="b")
x,y = getByClass(2,1,2)
pl.plot(x,y,'o',color="g")
pl.show()
```

# Ajustando via Naive Bayes

```python
gnb = GaussianNB()
model = gnb.fit(iris.data, iris.target)
```

```python
predict = model.predict(iris.data)
```

```python
print("Number of mislabeled points : %d" % (iris.target != predict).sum())
```
