```python
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)
z = np.random.normal(1.1,8, 1000)
```

```python
plt.figure()
# plt.hist(z,color='red',alpha=0.2, bins=100)
plt.hist(x,color='blue',alpha=0.2, bins=100)
plt.hist(z,color='green',alpha=0.2, bins=100)
plt.show()
```

```python
z_z = (z-np.mean(z))/np.std(z)
z_x = (x-np.mean(x))/np.std(x)
```

```python
plt.figure()
# plt.hist(z,color='red',alpha=0.2, bins=100)
plt.hist(z_x,color='blue',alpha=0.2, bins=100)
plt.hist(z_z,color='green',alpha=0.2, bins=100)
plt.show()
```

```python
count, bins = np.histogram(x, bins=100, normed=True)
plt.figure()
plt.plot(np.arange(len(count)),count,"r-")
plt.title(u'Count histogram !')
plt.show()
```

```python
np.linspace(-15, 15, 9)
```

```python
cdf  = []
total = 0
for i in count:
    total += i
    cdf.append(total)
    
# print(cdf)
plt.figure()
plt.plot(np.arange(len(cdf)),np.array(cdf)/len(cdf),"r-")
plt.title(u'CDF !')
plt.show()
```
