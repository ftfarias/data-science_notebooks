```python
import csv
import collections
input_file = open('/Users/developer/Projects/labs/hightower/busca_relacionadas/mapa_site.csv')
input_csv = csv.reader(input_file)
counter = collections.Counter()
for i,row in enumerate(input_csv):
    evento_anterior = row[3]
    evento_posterior = row[5]
    if evento_anterior in printable_nodes and evento_posterior in printable_nodes:
        if i > 100000:
            break
    #     print(evento_anterior,evento_exterior)
        counter.update([(evento_anterior,evento_posterior)])
print(counter.most_common())
```

```python
import collections
valores = counter.most_common()
valores_total = collections.defaultdict(float)
for v,e in valores:
    evento_anterior = v[0]
    evento_posterior = v[1]
    valores_total[evento_anterior] += e
valores_total
```

```python
from graphviz import Digraph

valores = counter.most_common()

dot = Digraph(comment='Mapa Site Elo7')

for v,e in valores:
#     print(v)
#     print(v[1])
    evento_anterior = v[0]
    evento_posterior = v[1]
    dot.node(evento_anterior, evento_anterior)
    dot.node(evento_posterior, evento_posterior)
    percent =  1.0*e/valores_total[evento_anterior]
    if percent > 0.05:
        percent_label = '{:.0%}'.format(percent)
        dot.edge(evento_anterior, evento_posterior, label=percent_label)
    
```

```python
dot.render('mapa_do_site', view=False)
dot
```

```python
# graph = counter.keys()
graph = [k[0] for k in counter.most_common(10)]

labels=None
graph_layout='spring'
node_size=1600
node_color='blue'
node_alpha=0.3
node_text_size=12
edge_color='blue'
edge_alpha=0.3
edge_tickness=1
edge_text_pos=0.3
text_font='sans-serif'

# create networkx graph
G=nx.DiGraph()

# add edges
for edge in graph:
    G.add_edge(edge[0], edge[1])

# these are different layouts for the network you may try
# shell seems to work best
if graph_layout == 'spring':
    graph_pos=nx.spring_layout(G)
elif graph_layout == 'spectral':
    graph_pos=nx.spectral_layout(G)
elif graph_layout == 'random':
    graph_pos=nx.random_layout(G)
else:
    graph_pos=nx.shell_layout(G)

# draw graph
nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, 
                       alpha=node_alpha, node_color=node_color)

nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,
                       alpha=edge_alpha,edge_color=edge_color)

nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                        font_family=text_font)

if labels is None:
    labels = range(len(graph))

edge_labels = dict(zip(graph, labels))

nx.draw_networkx_edge_labels(G, graph_pos, edge_labels=edge_labels, 
                             label_pos=edge_text_pos)

# show graph
plt.show()
```

```python


def draw_graph(graph, labels=None, graph_layout='shell',
               node_size=1600, node_color='blue', node_alpha=0.3,
               node_text_size=12,
               edge_color='blue', edge_alpha=0.3, edge_tickness=1,
               edge_text_pos=0.3,
               text_font='sans-serif'):

    # create networkx graph
    G=nx.Graph()

    # add edges
    for edge in graph:
        G.add_edge(edge[0], edge[1])

    # these are different layouts for the network you may try
    # shell seems to work best
    if graph_layout == 'spring':
        graph_pos=nx.spring_layout(G)
    elif graph_layout == 'spectral':
        graph_pos=nx.spectral_layout(G)
    elif graph_layout == 'random':
        graph_pos=nx.random_layout(G)
    else:
        graph_pos=nx.shell_layout(G)

    # draw graph
    nx.draw_networkx_nodes(G,graph_pos,node_size=node_size, 
                           alpha=node_alpha, node_color=node_color)
    nx.draw_networkx_edges(G,graph_pos,width=edge_tickness,
                           alpha=edge_alpha,edge_color=edge_color)
    nx.draw_networkx_labels(G, graph_pos,font_size=node_text_size,
                            font_family=text_font)

    if labels is None:
        labels = range(len(graph))

    edge_labels = dict(zip(graph, labels))
    nx.draw_networkx_edge_labels(G, graph_pos, edge_labels=edge_labels, 
                                 label_pos=edge_text_pos)

    # show graph
    plt.show()

graph = [(0, 1), (1, 5), (1, 7), (4, 5), (4, 8), (1, 6), (3, 7), (5, 9),
         (2, 4), (0, 4), (2, 5), (3, 6), (8, 9)]

# you may name your edge labels
labels = map(chr, range(65, 65+len(graph)))
draw_graph(graph, labels)

# if edge labels is not specified, numeric labels (0, 1, 2...) will be used
#draw_graph(graph)
```
