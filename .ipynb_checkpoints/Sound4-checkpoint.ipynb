```python
from itertools import count
import math
import random

def sine_wave(frequency=440.0, framerate=44000, amplitude=1.0):
    if amplitude > 1.0:
        amplitude = 1.0
    if amplitude < 0.0:
        amplitude = 0.0
    frequency = float(frequency)
    return (amplitude * math.sin(2.0 * math.pi * frequency * (float(i) / framerate)) for i in
            count(0))

def sine_wave_array(size, frequency=440.0, framerate=44000, amplitude=1.0):
    wave = sine_wave(frequency=frequency, framerate=framerate, amplitude=amplitude)
    return [wave.next() for _ in xrange(size)]

def wave_array(size, functions):
    norm = len(functions)
    return [sum((f.next() for f in functions))/norm for _ in xrange(size)]

def white_noise(amplitude=0.5):
    return (float(amplitude) * random.uniform(-1, 1) for _ in count(0))
```

```python
x1 = sine_wave(frequency=100, framerate=44000, amplitude=1.0)
x2 = sine_wave(frequency=400, framerate=44000, amplitude=1.0)
n = white_noise(amplitude=6)
x = wave_array(2048,[x1,x2,n])
plt.figure()
plt.plot(x)
```

```python
%time f = np.fft.rfft(x)
n = len(f)/2
fa = [abs(x)/n for x in f[1:len(f)/2]]
plt.figure()
plt.plot(fa)
```

```python
import math

def sum_of_decibels(db_lists):
    if db_lists == []:
        return 0
    powers = [10 ** (db/10) for db in db_lists ]
    total_power = sum(powers)
    return 10 * math.log10(total_power)

background_noise_100hz = 60.0
background_noise_1khz = 50.0

def n(freq):
    logfreq = math.log10(freq)
    base = [0]
    
     
    if freq < 50:
        base.append(120 - 60 * logfreq)    
    
#   shipping
    a = 20.0
    h = 1.5
    if 1 < freq < 10000:  # 10  -  100
        v = 90 - (a* ((logfreq - h)**2))
        base.append(v)    
    
    # sea
    Sea_noise_ = 90
    a = 50.0
    h = 2.7
    if 50 < freq < 1000: # 100 - 1000    
        v = Sea_noise_ - (a* (((logfreq) - h)**2))
        base.append(v)   
    
    if 1000 < freq < 1000000: # 100 - 1000
        noise_1khz = Sea_noise_ - (a* ((3 - h)**2))
        base.append(noise_1khz - (17.0 * (logfreq-3)))

    
    if 1000 < freq:
        base.append(-75 + 20 * logfreq)

    return sum_of_decibels(base) + random.gauss(0, 1)
```

```python
x = np.logspace(0,7,num=1000)
y = []
for v in x:
    y.append(n(v))
plt.figure()
#plt.axes[0].set_ylim(0,100)
plt.semilogx(x,y)
```
