Nesta etapa muita coisa é feita. Para certos valores nulos eu posso tratar todo o dataset igualmente. Mas para utilização da mediana eu preciso antes criar meus grupos de treino e teste e tratar as features de teste utilizando os valores da mediana do treino:

```python
selected_features.loc[selected_features.VLR_FATURADO_ANTERIOR.isnull(),'VLR_FATURADO_ANTERIOR'] = 0
selected_features.loc[selected_features.QTD_RENOVACAO.isnull(),'QTD_RENOVACAO'] = 0
selected_features.loc[selected_features.QTD_PRE_RENOVE.isnull(),'QTD_PRE_RENOVE'] = 0
selected_features.loc[selected_features.QTD_PARCELA_ORIGEM.isnull(),'QTD_PARCELA_ORIGEM'] = 0
selected_features.loc[selected_features.TEMPO_PRE_RENOV_MEDIO_DIAS.isnull(),'TEMPO_PRE_RENOV_MEDIO_DIAS'] = 0

selected_features.loc[selected_features.VIGENCIA_EDICOES_ORIGEM.isnull(),'VIGENCIA_EDICOES_ORIGEM'] = 0
selected_features.loc[selected_features.VIGENCIA_EDICOES_ANTERIOR.isnull(),'VIGENCIA_EDICOES_ANTERIOR'] = 0
selected_features.loc[selected_features.VIGENCIA_MESES_ANTERIOR.isnull(),'VIGENCIA_MESES_ANTERIOR'] = 0
selected_features.loc[selected_features.VIGENCIA_EDICOES_ORIGEM_ANTERIOR.isnull(),'VIGENCIA_EDICOES_ORIGEM_ANTERIOR'] = 0
selected_features.loc[selected_features.VIGENCIA_MESES_ORIGEM_ANTERIOR.isnull(),'VIGENCIA_MESES_ORIGEM_ANTERIOR'] = 0

selected_features.loc[selected_features.DESCONTO_ANTERIOR.isnull(),'DESCONTO_ANTERIOR'] = 0

# load dos ids de treino e teste
ids_treino = []
for line in open('../data/treino_ids.txt'):
    ids_treino.append(int(line.strip('\n').split(',')[0]))
ids_teste = []
for line in open('../data/test_ids.txt'):
    ids_teste.append(int(line.strip('\n').split(',')[0]))

ids_treino.sort()
ids_teste.sort()

selected_features.loc[selected_features.IDADE.isnull(),'IDADE'] = selected_features.loc[ids_treino,'IDADE'].median()
selected_features.loc[selected_features.DESCONTO.isnull(),'DESCONTO'] = selected_features.loc[ids_treino,'DESCONTO'].median()
selected_features.loc[selected_features.per_h_resid_10.isnull(),'per_h_resid_10'] = selected_features.loc[ids_treino,'per_h_resid_10'].median()
selected_features.loc[selected_features.perc_m_resd_10.isnull(),'perc_m_resd_10'] = selected_features.loc[ids_treino,'perc_m_resd_10'].median()
selected_features.loc[selected_features.perc_alf_10.isnull(),'perc_alf_10'] = selected_features.loc[ids_treino,'perc_alf_10'].median()
selected_features.loc[selected_features.renda_sm_2010.isnull(),'renda_sm_2010'] = selected_features.loc[ids_treino,'renda_sm_2010'].median()
selected_features.loc[selected_features.GINI.isnull(),'GINI'] = selected_features.loc[ids_treino,'GINI'].median()
selected_features.loc[selected_features.IDHM.isnull(),'IDHM'] = selected_features.loc[ids_treino,'IDHM'].median()
selected_features.loc[selected_features.IDHM_E.isnull(),'IDHM_E'] = selected_features.loc[ids_treino,'IDHM_E'].median()
selected_features.loc[selected_features.IDHM_L.isnull(),'IDHM_L'] = selected_features.loc[ids_treino,'IDHM_L'].median()
selected_features.loc[selected_features.IDHM_R.isnull(),'IDHM_R'] = selected_features.loc[ids_treino,'IDHM_R'].median()

selected_features.loc[selected_features.PIB.isnull(),'PIB'] = selected_features.loc[ids_treino,'PIB'].median()
selected_features.loc[selected_features.Populacao.isnull(),'Populacao'] = selected_features.loc[ids_treino,'Populacao'].median()
selected_features.loc[selected_features.PIBPerCapita.isnull(),'PIBPerCapita'] = selected_features.loc[ids_treino,'PIBPerCapita'].median()

selected_features.loc[selected_features.qtd_dom.isnull(),'qtd_dom'] = selected_features.loc[ids_treino,'qtd_dom'].median()
selected_features.loc[selected_features.qtd_mor.isnull(),'qtd_mor'] = selected_features.loc[ids_treino,'qtd_mor'].median()
selected_features.loc[selected_features.med_mor.isnull(),'med_mor'] = selected_features.loc[ids_treino,'med_mor'].median()


X_treino = selected_features.iloc[ids_treino]
X_test = selected_features.iloc[ids_teste]
print "Dimensionalidade antes de remover linhas problemas"
print X_treino.shape
print X_test.shape
X_treino = X_treino[X_treino['NUM_PERIODO'] < 900]
X_test = X_test[X_test['NUM_PERIODO'] < 900]
print "Dimensionalidade depois remover linhas problemas"
print X_treino.shape
print X_test.shape
# coluna da variavel resposta
y_train = X_treino.CHURN.as_matrix()
y_test = X_test.CHURN.as_matrix()

# save para bigml
#X_test.to_csv("test_bigml.csv", index=False)

del X_treino['CHURN']
del X_test['CHURN']
print "Dimensionalidade depois remover coluna CHURN"
print X_treino.shape
print X_test.shape
```

Nesta etapa eu treino o modelo e classifico o conjunto de teste. O resultado da classificação é apresentado ao final.

```python
# Testing

from sklearn.ensemble import GradientBoostingClassifier
#from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

# Training GBM with best Parameters:

est = GradientBoostingClassifier(n_estimators=1000, max_depth=6, learning_rate=0.1)
est.fit(X_treino.values, y_train)

pred = est.predict(X_test.values)

from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
from sklearn.metrics import roc_auc_score
print(roc_auc_score (y_test, pred))
```
