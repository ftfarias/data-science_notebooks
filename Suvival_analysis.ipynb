```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas.io.pytables import HDFStore
from lifelines.plotting import plot_lifetimes
from numpy.random import uniform, exponential

# http://lifelines.readthedocs.org/en/latest/Survival%20Analysis%20intro.html

N = 25
current_time = 10
actual_lifetimes = np.array([[exponential(12), exponential(2)][uniform()<0.5] for i in range(N)])
observed_lifetimes = np.minimum(actual_lifetimes,current_time)
observed= actual_lifetimes < current_time

plt.xlim(0,25)
plt.vlines(10,0,30,lw=2, linestyles="--")
plt.xlabel('time')
plt.title('Births and deaths of our population, at $t=10$')
plot_lifetimes(observed_lifetimes, event_observed=observed)
print "Observed lifetimes at time %d:\n"%(current_time), observed_lifetimes
```

# Non-parametrically fit the survival curve:

```python
from lifelines import KaplanMeierFitter

kmf = KaplanMeierFitter()
kmf.fit(observed_lifetimes, event_observed=observed)


# fitter methods have an internal plotting method.
# plot the curve with the confidence intervals
kmf.plot()
```

# Non-parametrically fit the cumulative hazard curve:

```python
from lifelines import NelsonAalenFitter

naf = NelsonAalenFitter()
naf.fit(observed_lifetimes, event_observed=observed) 

#plot the curve with the confidence intervals
naf.plot()
```
