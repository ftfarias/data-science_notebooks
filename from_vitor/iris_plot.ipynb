## Iris

```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
```

```python
data = load_iris()
features = data['data']
features_names = data['feature_names']
target = data['target']
target_names = data['target_names']
labels = target_names[target]
```

```python
from itertools import combinations
features_indexes = range(len(features[0]))
pairs_of_features = [tuple(c) for c in combinations(features_indexes,2)]
for i, (p0, p1) in enumerate(pairs_of_features):
    plt.subplot(2,3,i+1)
    for t, marker, c in zip(xrange(3), ">ox", "rgb"):
        plt.scatter(features[target == t, p0], features[target == t, p1], marker=marker, c=c)
    plt.xlabel(features_names[p0])
    plt.ylabel(features_names[p1])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
plength = features[:, 2]
is_setosa = (labels == 'setosa')
max_setosa = plength[is_setosa].max()
min_non_setosa = plength[~is_setosa].min()
print("Maximum petal length of a setosa: {0}".format(max_setosa))
print("Minimum petal length of a non-setosa: {0}".format(min_non_setosa))
```
