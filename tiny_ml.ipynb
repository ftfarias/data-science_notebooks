# Carregando os dados

```python
%matplotlib inline
import scipy as sp
data = sp.genfromtxt("web_traffic.tsv", delimiter="\t")
print(data[:10])
print data.shape
```

# Pre-processamento e limpeza dos dados

```python
x = data[:,0]
y = data[:,1]
sp.sum(sp.isnan(y))
```

```python
x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,alpha=0.3)
plt.title("Tráfego no último mês".decode('utf-8'))
plt.xlabel("Tempo")
plt.ylabel("Hits/Hora")
plt.xticks([w*7*24 for w in range(10)], ['week %i' %w for w in range(10)])
plt.autoscale(tight=True)
plt.grid()
plt.show()
```

# Escolhendo modelo e algoritmo de aprendizado

```python
def error(f, x, y):
    return sp.sum((f(x) - y)**2)
```

## Modelo Simples

```python
fp1, residuals, rank, sv, rcond = sp.polyfit(x, y, 1, full=True)
print("Parâmetros do modelo: %s" % fp1)
print("Erro da aproximação: %f" % residuals)
```

```python
print("Isso significa que nossa reta tem a seguinte forma:")
print("f(x) = %f * x + %f" % (fp1[0], fp1[1]))
```

```python
f1 = sp.poly1d(fp1)
print(error(f1,x,y))
```

```python
from plotmodels import plot_models
```

```python
plot_models(x, y, [f1])
```

## Modelo um pouco mais complicado

```python
f2p = sp.polyfit(x, y, 2)
print(f2p)
f2 = sp.poly1d(f2p)
print(error(f2, x, y))
```

```python
print("Isso significa que nossa reta tem a seguinte forma:")
print("f(x) = %f * x**2 %f * x + %f" % (f2p[0], f2p[1], f2p[2]))
```

```python
plot_models(x, y, [f1, f2])
```

## Vamos complicar as coisas

```python
f3 = sp.poly1d(sp.polyfit(x, y, 3))
f10 = sp.poly1d(sp.polyfit(x, y, 10))
```

```python
plot_models(x, y, [f1, f2, f3, f10])
```

```python
print("Erro f1: " + str(error(f1, x, y)))
print("Erro f2: " + str(error(f2, x, y)))
print("Erro f3: " + str(error(f3, x, y)))
print("Erro f10: " + str(error(f10, x, y)))
```

## Podemos fazer melhor?

```python
inflection = 3.5*7*24
xa = x[:inflection]
ya = y[:inflection]
xb = x[inflection:]
yb = y[inflection:]

fa = sp.poly1d(sp.polyfit(xa, ya, 1))
fb = sp.poly1d(sp.polyfit(xb, yb, 1))
fa_error = error(fa, xa, ya)
fb_error = error(fb, xb, yb)
print("Error inflexão = %f" % (fa_error + fb_error))
```

```python
plot_models(x, y, [fa, fb])
```

# E o que o futuro nos aguarda?

```python
plot_models(x, y, [f1, f2, f3, f10], mx=sp.linspace(0 * 7 * 24, 6 * 7 * 24, 100), ymax=10000, xmin=0 * 7 * 24)
```

```python
print("Treino com dados gerados depois do ponto de inflexão")
fb1 = fb
fb2 = sp.poly1d(sp.polyfit(xb, yb, 2))
fb3 = sp.poly1d(sp.polyfit(xb, yb, 3))
fb10 = sp.poly1d(sp.polyfit(xb, yb, 10))
```

```python
print("Erros com dados gerados depois do ponto de inflexão")
for f in [fb1, fb2, fb3, fb10]:
    print("Error d=%i: %f" % (f.order, error(f, xb, yb)))
```

```python
plot_models(x, y, [fb1, fb2, fb3, fb10], mx=sp.linspace(0 * 7 * 24, 6 * 7 * 24, 100), ymax=10000, xmin=0 * 7 * 24)
```

# Separando treino e teste

```python
frac = 0.3
split_idx = int(frac * len(xb))
shuffled = sp.random.permutation(list(range(len(xb))))
test = sorted(shuffled[:split_idx])
train = sorted(shuffled[split_idx:])
fbt1 = sp.poly1d(sp.polyfit(xb[train], yb[train], 1))
fbt2 = sp.poly1d(sp.polyfit(xb[train], yb[train], 2))
fbt3 = sp.poly1d(sp.polyfit(xb[train], yb[train], 3))
fbt10 = sp.poly1d(sp.polyfit(xb[train], yb[train], 10))
```

```python
print("Erros no teste para dados gerados depois do ponto de inflexão:")
for f in [fbt1, fbt2, fbt3, fbt10]:
    print("Error d=%i: %f" % (f.order, error(f, xb[test], yb[test])))
```

```python
plot_models(x, y, [fbt1, fbt2, fbt3, fbt10], mx=sp.linspace(0 * 7 * 24, 6 * 7 * 24, 100), ymax=10000, xmin=0 * 7 * 24)
```
